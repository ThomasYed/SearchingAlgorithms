```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("datagraph.txt") as file:
  lines = [line.split() for line in file]
with open("mmapdata.txt") as file:
  mmapl = [line.split() for line in file]
```

```python
x = np.array(lines[0]).astype(int)
y1 = np.array(lines[1]).astype(float)
y2 = np.array(lines[2]).astype(float)
y3 = np.array(lines[3]).astype(float)
y4 = np.array(mmapl[0]).astype(float)
```

```python
plt.figure(figsize=(10, 8), dpi=100)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(['Binary Tree', 'Hash Table', 'RedBlack Tree'])
plt.title("Зависимость времени поиска 100 элементов от объёма данных")
plt.show()
```
